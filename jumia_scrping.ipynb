```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
driver = webdriver.Chrome() 
BASE_URL = "https://www.jumia.com.ng/phones-tablets/"
products, prices, ratings, images = [], [], [], []
for page in range(1, 4):
    print(f"Scraping page {page}...")
    url = f"{BASE_URL}?page={page}"
    driver.get(url)
    time.sleep(5) 
    items = driver.find_elements(By.CLASS_NAME, "prd")
    print("Items found:", len(items))
    for item in items:
        try:
            name = item.find_element(By.CLASS_NAME, "name").text
        except:
            name = "N/A"
        try:
            price = item.find_element(By.CLASS_NAME, "prc").text
        except:
            price = "N/A"
        try:
            rating = item.find_element(By.CLASS_NAME, "stars").text
        except:
            rating = "No Rating"
        try:
            image = item.find_element(By.TAG_NAME, "img").get_attribute("src")
        except:
            image = "N/A"
        products.append(name)
        prices.append(price)
        ratings.append(rating)
        images.append(image)
driver.quit()
df = pd.DataFrame({
    "Product": products,
    "Price": prices,
    "Rating": ratings,
    "Image": images
})
df 
```
